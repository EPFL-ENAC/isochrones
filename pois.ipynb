```python
from matplotlib import pyplot as plt
from matplotlib import colors as colors

from isochrones.pois import get_osm_features
```

```python
bounding_box = (
    5.882835,
    46.068877,
    6.337395,
    46.325682,
)  # from bbxofinder.com, Collonges - Fillinges, Cessy - La roche sur Foron

tags = {
    "amenity": True,
    "healthcare": True,
    "office": True,
    "public_Transport": True,
    "shop": True,
    "tourism": True,
}
```

```python
pois = get_osm_features(bounding_box, tags)
```

```python
pois
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

pois.plot(
    ax=ax, column="variable", categorical=True, legend=True, markersize=5, alpha=0.7
)
```
